```python
x=df.iloc[:,[1,2,3]].values
print(x)
```

<h3>find optimum number of clusters</h3>

```python
a=[]
for i in range(1,20):
    model=KMeans(i)
    model.fit(x)
    a.append(model.inertia_)
```

```python
a
```

<h3>plot line to show elbow</h3>

```python
plt.style.use('fivethirtyeight')
plt.plot(range(1,20),a)
plt.title("Elbow method")
plt.xlabel("number of clusters")
plt.ylabel("within clusters sum of squares")
```

The elbow is formed at 3 so the optimum number of clusters is 3

<h3>apply k-means to dataset</h3>

```python
model=KMeans(3)
y_kmeans=model.fit_predict(x)
print(y_kmeans)
```

<h3>visualize clusters</h3>

```python
plt.style.use("classic")
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='Iris-setosa')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='yellow',label='Iris-versicolour')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='blue',label='Iris-virginica')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=100,c='green',label='centroid')
plt.legend()
plt.show()
```
